```python
import qiskit 
qiskit.__qiskit_version__
from qiskit import *
from qiskit.visualization import plot_histogram
#IBMQ.save_account('')
IBMQ.load_account()
```

```python
qr = QuantumRegister(2)
```

```python
cr = ClassicalRegister(2)
```

```python
c = QuantumCircuit(qr, cr)
```

```python
%matplotlib inline
c.draw()
```

```python
c.h(qr[0])
c.draw(output="mpl")
```

```python
c.cx(qr[0], qr[1])
c.draw(output="mpl")
```

```python
c.measure(qr, cr)
c.draw(output="mpl")
```

```python
simulator = Aer.get_backend("qasm_simulator")
result = execute(c, backend = simulator).result()
```

```python
plot_histogram(result.get_counts(c))
```

```python
provider = IBMQ.get_provider(hub='ibm-q', group='open', project='main')
provider.backends()
```

```python
qcomp = provider.get_backend('ibmq_bogota')
```

```python
job = execute(c, backend=qcomp)
```
